### 1. Sales Overview:To Understand the overall sales performance, trends, and patterns over time.

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y',errors="coerce")
```

```python
df = df.dropna(subset=['Date'])
```

```python
sales_by_date = df.groupby('Date').agg({
    'Order ID': 'count',
    'Amount': 'sum'
}).rename(columns={'Order ID': 'Order Count', 'Amount': 'Total Sales Amount'})
fig, ax1 = plt.subplots(figsize=(14, 7))
ax1.set_xlabel('Date')
ax1.set_ylabel('Order Count', color='tab:blue')
ax1.plot(sales_by_date.index, sales_by_date['Order Count'], color='tab:blue', label='Order Count')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax2 = ax1.twinx()
ax2.set_ylabel('Total Sales Amount (INR)', color='tab:orange')
ax2.plot(sales_by_date.index, sales_by_date['Total Sales Amount'], color='tab:orange', label='Total Sales Amount')
ax2.tick_params(axis='y', labelcolor='tab:orange')
fig.tight_layout()
plt.title('Sales Trends Over Time')
plt.show()
```

This shows Peak Sales in the month of April.

```python
print(df1['Date'].dtype)

df1['Date'] = pd.to_datetime(df1['Date'], errors='coerce')

num_nat = df1['Date'].isna().sum()
print(f"Number of NaT values: {num_nat}")

df1.dropna(subset=['Date'], inplace=True)

df1['Year'] = df1['Date'].dt.year
df1['Month'] = df1['Date'].dt.month
df1['Day'] = df1['Date'].dt.day
df1['Week'] = df1['Date'].dt.isocalendar().week
df1['Month-Year'] = df1['Date'].dt.to_period('M')

monthly_sales = df1.groupby('Month-Year').agg({'Amount': 'sum', 'Order ID': 'count'}).rename(columns={'Amount': 'Total Sales', 'Order ID': 'Total Orders'})
monthly_sales['Average Order Value'] = monthly_sales['Total Sales'] / monthly_sales['Total Orders']

monthly_sales['Monthly Sales Growth'] = monthly_sales['Total Sales'].pct_change()

plt.figure(figsize=(14, 7))

# Total Sales Over Time
plt.subplot(2, 2, 1)
monthly_sales['Total Sales'].plot()
plt.title('Total Sales Over Time')
plt.xlabel('Month-Year')
plt.ylabel('Total Sales')

# Number of Orders Over Time
plt.subplot(2, 2, 2)
monthly_sales['Total Orders'].plot()
plt.title('Number of Orders Over Time')
plt.xlabel('Month-Year')
plt.ylabel('Total Orders')

# Average Order Value Over Time
plt.subplot(2, 2, 3)
monthly_sales['Average Order Value'].plot()
plt.title('Average Order Value Over Time')
plt.xlabel('Month-Year')
plt.ylabel('Average Order Value')

# Sales Growth Rate Over Time
plt.subplot(2, 2, 4)
monthly_sales['Monthly Sales Growth'].plot()
plt.title('Monthly Sales Growth Rate')
plt.xlabel('Month-Year')
plt.ylabel('Growth Rate')

plt.tight_layout()
plt.show()

# Calculate specific insights
peak_sales_month = monthly_sales['Total Sales'].idxmax()
peak_sales_value = monthly_sales['Total Sales'].max()

average_sales_per_month = monthly_sales['Total Sales'].mean()
average_orders_per_month = monthly_sales['Total Orders'].mean()

print(f"Peak sales month: {peak_sales_month}")
print(f"Peak sales value: {peak_sales_value:.2f}")
print(f"Average sales per month: {average_sales_per_month:.2f}")
print(f"Average number of orders per month: {average_orders_per_month:.2f}")

high_growth_months = monthly_sales[monthly_sales['Monthly Sales Growth'] > 0.20]
print("Months with more than 20% sales growth:")
print(high_growth_months[['Total Sales', 'Monthly Sales Growth']])
```

```python
fig, ax = plt.subplots(9, 1, figsize=(15, 50))

cat = df1['Category'].unique().tolist()
cat = iter(cat)

for i in range(9):
    a = next(cat)
    df = df1[df1['Category'] == a].copy()
    sns.lineplot(x='Date', y='Amount', data=df, ax=ax[i])
    ax[i].set_title(f'Sales of {a}', fontsize=16)
    ax[i].set_xlabel('Date', fontsize=12)
    ax[i].set_ylabel('Amount', fontsize=12)
    ax[i].grid(True)

plt.tight_layout()
plt.show()
```

### 2. Product Analysis: Analyze the distribution of product categories, sizes, and quantities sold to identify popular products.

```python
category_orders = df1['Category'].value_counts()
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")  

ax = sns.barplot(x=category_orders.index, y=category_orders.values, palette="viridis")
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 9), 
                textcoords='offset points',
                fontsize=10, color='black')  
plt.title('Number of Orders by Product Category', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Number of Orders', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12) 
plt.yticks(fontsize=12)  
plt.tight_layout()
plt.show()
```

This graph shows the number of orders for different categories, where Tshirt has the maximum number of orders and Watch has 
the minimum number of orders.

```python
size_distribution = df1['Size'].value_counts()
plt.figure(figsize=(14, 7))
sns.set_style("whitegrid")  
ax = sns.barplot(x=size_distribution.index, y=size_distribution.values, palette="viridis")
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 9), 
                textcoords='offset points',
                fontsize=10, color='black')  

plt.title('Product Size Distribution', fontsize=16)
plt.xlabel('Size', fontsize=14)
plt.ylabel('Number of Orders', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)  
plt.yticks(fontsize=12)  
plt.tight_layout()
plt.show()
```

This graph shows the distribution of product size, where M has maximum number of orders and 4XL has minimum.

```python
category_amount = df1.groupby("Category")["Amount"].sum().sort_values(ascending=False)


plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")  
ax = sns.barplot(x=category_amount.index, y=category_amount.values, palette="viridis")
for p in ax.patches:
    ax.annotate("{:,.2f}".format(p.get_height()), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 9), 
                textcoords='offset points',
                fontsize=10, color='black')  

plt.title('Total Amount Earned by Product Category', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Total Amount ', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)  
plt.yticks(fontsize=12)  
plt.tight_layout()
plt.show()
```

This shows the total amount earned by different categories where T-shirt has maximum Sales Amount.

```python
category_amount = df1.groupby("Category")["Qty"].sum().reset_index()
category_amount_sorted = category_amount.sort_values(by='Qty', ascending=False)
print(category_amount_sorted)
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x="Category", y="Qty", data=category_amount_sorted, palette="viridis")
bar_plot.set_title('Total Quantity by Category', fontsize=16)
bar_plot.set_xlabel('Category', fontsize=14)
bar_plot.set_ylabel('Total Quantity', fontsize=14)
plt.xticks(rotation=45)
for container in bar_plot.containers:
    bar_plot.bar_label(container)
plt.tight_layout()
plt.show()
```

This graph shows T-shirt has maximum quantity So we can focus on the Marketing strategies and Product Diversification for Tshirt

```python
monthly_category_amount = df1.groupby(['Month', 'Category'])['Qty'].sum().reset_index()
most_popular_by_month = monthly_category_amount.loc[monthly_category_amount.groupby('Month')['Qty'].idxmax()]

print(most_popular_by_month)
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
bar_plot = sns.barplot(x="Month", y="Qty", hue="Category", data=most_popular_by_month, palette="viridis")
bar_plot.set_title('Most Popular Product by Month', fontsize=16)
bar_plot.set_xlabel('Month', fontsize=14)
bar_plot.set_ylabel('Total Quantity', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
customer_segments = df1.groupby('Order ID').agg({
    'Amount': 'sum',
    'Qty': 'sum'
}).rename(columns={'Amount': 'total_spent', 'Qty': 'order_count'})

# Visualize customer segmentation
plt.figure(figsize=(12, 6))
plt.scatter(customer_segments['order_count'], customer_segments['total_spent'])
plt.title('Customer Segmentation')
plt.xlabel('Number of Orders')
plt.ylabel('Total Amount Spent')
plt.grid(True)
plt.show()
```

```python
city_customers = df1.groupby('ship-city').size().reset_index(name='Customer Count')
top_cities_customers = city_customers.sort_values(by='Customer Count', ascending=False).head(10)
plt.figure(figsize=(14, 8))
sns.set(style="whitegrid")
barplot = sns.barplot(y='ship-city', x='Customer Count', data=top_cities_customers, palette='viridis')
for container in barplot.containers:
    barplot.bar_label(container, fmt='%d', label_type='edge', fontsize=12)
plt.ylabel("City", fontsize=16)
plt.xlabel("Number of Customers", fontsize=16)
plt.title("Top 10 Cities by Number of Customers", fontsize=20, pad=20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.show()
```

```python
state_purchase_counts = df1.groupby('ship-state').size().reset_index(name='Total Purchases')
top_10_states = state_purchase_counts.sort_values(by='Total Purchases', ascending=False).head(10)
top_10_states_df = df1[df1['ship-state'].isin(top_10_states['ship-state'])]
popular_categories = top_10_states_df.groupby(['ship-state', 'Category']).size().reset_index(name='Counts')
most_popular_categories = popular_categories.loc[popular_categories.groupby('ship-state')['Counts'].idxmax()]
plt.figure(figsize=(14, 10))
sns.set(style="whitegrid")
barplot = sns.barplot(y='ship-state', x='Counts', hue='Category', data=most_popular_categories, palette='viridis')
for container in barplot.containers:
    barplot.bar_label(container, fmt='%d', label_type='edge', fontsize=12)
plt.ylabel("State", fontsize=16)
plt.xlabel("Number of Purchases", fontsize=16)
plt.title("Most Popular Product Categories in Top 10 States by Purchases", fontsize=20, pad=20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(title='Category', title_fontsize='13', fontsize='11', loc='upper right')
plt.tight_layout()
plt.show()
```
